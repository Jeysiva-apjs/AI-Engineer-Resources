```python
import pandas as pd
```

```python
df_sales = pd.read_excel("linechart.xlsx")
```

```python
df_sales.head(3)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'], df_sales['Fridge'], label="Fridge")
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'], label="Dishwasher")
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'], label="Washing Machine")

plt.title("Product Sales")
plt.xlabel('Financial Quarter')
plt.ylabel('Revenue (mln $)')

plt.legend()

plt.show()
```

```python
sales_total = df_sales[["Fridge",  "Dishwasher",  "Washing Machine"]].sum()
sales_total
```

```python
plt.pie(sales_total, labels=sales_total.index, autopct="%1.1f%%", shadow=False, explode=(0.1,0,0), startangle=140)

plt.show()
```

```python
df_sales.plot(kind="bar", x="Quarter")

plt.xticks(rotation=45)

plt.show()
```

```python
df = df_sales.set_index("Quarter")

df.plot(kind="line")
plt.ylabel("Revenue in mln$")

plt.show()
```
